```python
import requests

# Replace 'your_api_endpoint_here' with the actual API endpoint URL for player data
api_url = 'https://fantasy.premierleague.com/api/bootstrap-static/'

# Make a GET request to the FPL API
response = requests.get(api_url)

if response.status_code == 200:
    data = response.json()
    
    # Extract and print player information
    players = data['elements']
    
    for player in players:
        player_name = f"{player['first_name']} {player['second_name']}"
        player_position = player['element_type']  # Position ID (e.g., 1 for Goalkeeper)
        
        # You can map position IDs to position names if needed
        positions = {
            1: 'Goalkeeper',
            2: 'Defender',
            3: 'Midfielder',
            4: 'Forward'
        }
        
        position_name = positions.get(player_position)
        
        print(f"Player Name: {player_name}, Position: {position_name}")
else:
    print("Failed to retrieve player data.")
```
