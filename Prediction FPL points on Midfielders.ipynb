```python
MID = ordered_fbref_data[ (ordered_fbref_data['Pos'] == "RW") | (ordered_fbref_data['Pos'] == "RM") | (ordered_fbref_data['Pos'] == "DM") | (ordered_fbref_data['Pos'] == "LW") | (ordered_fbref_data['Pos'] == "LM") | (ordered_fbref_data['Pos'] == "CM") | (ordered_fbref_data['Pos'] == "AM")]
```

```python
MID['Xp'] = (MID['npxG'] * 5) + ( MID['xAG'] * 3)  + 1 - (MID['CrdY']) # FPL Scoring System
```

```python
MID.loc[MID['Min'] > 60, 'Xp'] += 1 #adding a + 1 to players who have played >60 minutes
MID.loc[(MID['xGA'] < 1) & (MID['Min'] > 60) , 'Xp'] += 1 #Adding a + 1 to players who have less than a 1 XG and has played more than 60 minutes
```

```python
train_data_encoded_imputed = MID.fillna(MID.mean())
```

```python
train_data_encoded_imputed
```

```python
target = train_data_encoded_imputed['Xp']
```

```python
X = train_data_encoded_imputed[['name' , 'Min', 'Sh', 'SoT', 'Touches', 'Tkl', 'Int', 'Blocks', 'xG_x',
       'npxG', 'xAG', 'SCA', 'GCA', 'Cmp', 'Att', 'Cmp%', 'PrgP', 'Carries',
       'PrgC', 'Att.1', 'Succ', 'xGA']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, target, test_size=0.3, random_state=42)
```

```python

X_test_name = X_test['name']
X_test = X_test[[ 'Min', 'Sh', 'SoT', 'Touches', 'Tkl', 'Int', 'Blocks', 'xG_x',
       'npxG', 'xAG', 'SCA', 'GCA', 'Cmp', 'Att', 'Cmp%', 'PrgP', 'Carries',
       'PrgC', 'Att.1', 'Succ', 'xGA']]

X_train = X_train[[ 'Min', 'Sh', 'SoT', 'Touches', 'Tkl', 'Int', 'Blocks', 'xG_x',
       'npxG', 'xAG', 'SCA', 'GCA', 'Cmp', 'Att', 'Cmp%', 'PrgP', 'Carries',
       'PrgC', 'Att.1', 'Succ', 'xGA']]

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

# Create a Random Forest Regressor
regressor = RandomForestRegressor(random_state=42)
# Train the regressor on the training data
regressor.fit(X_train, y_train)
# Make predictions on the test data
y_pred = regressor.predict(X_test)
# Evaluate the regressor using a regression metric (e.g., Mean Squared Error)
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```
