<h3> Model Testing </h3> 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, target, test_size=0.3, random_state=42)
X_train2, X_test2, y_train2, y_test2 = train_test_split(X, target2, test_size=0.3, random_state=42)
```

```python
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)

print('-------------------')
print("X_train shape:", X_train2.shape)
print("X_test shape:", X_test2.shape)
print("y_train shape:", y_train2.shape)
print("y_test shape:", y_test2.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

# Creating a Random Forest Regressor
regressor = RandomForestRegressor(random_state=42)
regressor2 = RandomForestRegressor(random_state= 42)
# Train the regressor on the training data
regressor.fit(X_train, y_train)
regressor2.fit(X_train2, y_train2)
# Make predictions on the test data
y_pred = regressor.predict(X_test)
y_pred2 = regressor2.predict(X_test2)
# Evaluate the regressor using a regression metric (e.g., Mean Squared Error)
mse = mean_squared_error(y_test, y_pred)
mse2 = mean_squared_error(y_test2, y_pred2 )
print("Mean Squared Error:", mse)
print("Second Mean Squared Error: ", mse2)

```
